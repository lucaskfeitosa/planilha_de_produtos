```python
import tabula
import pandas as pd


pdf_path = "Produtos.pdf"


tables = tabula.read_pdf(pdf_path, pages="all", multiple_tables=True)


result_dfs = []


for i, df in enumerate(tables):
    
    columns_to_extract = ['Codigo', 'Nome', 'Custo']
    if all(col in df.columns for col in columns_to_extract):
        
        result_df = df[columns_to_extract]
        result_dfs.append(result_df)
        print(f"Tabela {i+1}: Dados extraídos com sucesso.")
    else:
        print(f"Tabela {i+1}: Não contém as colunas necessárias.")


if result_dfs:
    final_result_df = pd.concat(result_dfs, ignore_index=True)

    
    output_file = "dados_extraidos.xlsx"
    final_result_df.to_excel(output_file, index=False)

    print(f"Todos os dados extraídos e salvos em {output_file}.")
else:
    print("Nenhuma tabela com as colunas necessárias foi encontrada.")
```
